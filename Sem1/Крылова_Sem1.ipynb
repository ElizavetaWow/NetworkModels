# Визуализация графов

## NetworkX

```python
import random
import networkx as nx
```

```python
def create_nodes(graph, n):
    graph.add_nodes_from([i for i in range(n)])
```

```python
def create_edges(graph, n=10):
    nodes = list(graph.nodes())
    nodes_lst = []
    if n > len(nodes)*(len(nodes)-1)/2:
        n = len(nodes)*(len(nodes)-1)/2
    while len(nodes_lst) != n:
        nodes1 = nodes.copy()
        i1 = random.choice(nodes1)
        nodes1.pop(i1)
        i2 = random.choice(nodes1)
        if (i1, i2) not in nodes_lst and (i2, i1) not in nodes_lst:
            nodes_lst.append((i1, i2))
    graph.add_edges_from(nodes_lst)
```

```python
def draw(graph):
    nx.draw_circular(graph,
         node_color='orange',
         node_size=500,
         with_labels=True)
```

```python
graph1 = nx.Graph()
create_nodes(graph1, 10)
create_edges(graph1, 20)
draw(graph1)
```

```python
graph2 = nx.Graph()
create_nodes(graph2, 50)
create_edges(graph2, 250)
draw(graph2)
```

```python
graph3 = nx.Graph()
create_nodes(graph3, 100)
create_edges(graph3, 200)
draw(graph3)
```

## Pyvis

```python
from pyvis.network import Network
```

```python
def create_nodes_pyvis(graph, n):
    graph.add_nodes([i for i in range(n)], 
    value=[random.randrange(1, 20, 2) for _ in range(n)], 
    color=["#"+''.join(random.choices('0123456789ABCDEF', k=6)) for _ in range(n)])
```

```python
def create_edges_pyvis(graph, n=10):
    nodes = graph.get_nodes()
    nodes_lst = []
    if n > len(nodes)*(len(nodes)-1)/2:
        n = len(nodes)*(len(nodes)-1)/2
    while len(nodes_lst) != n:
        nodes1 = nodes.copy()
        i1 = random.choice(nodes1)
        nodes1.pop(i1)
        i2 = random.choice(nodes1)
        if (i1, i2) not in nodes_lst and (i2, i1) not in nodes_lst:
            nodes_lst.append((i1, i2))
    graph.add_edges(nodes_lst)
```

```python
net1 = Network(notebook=True)
create_nodes_pyvis(net1, 10)
create_edges_pyvis(net1, 30)
net1.show('net1.html')
```

```python
net2 = Network(notebook=True)
create_nodes_pyvis(net2, 50)
create_edges_pyvis(net2, 100)
net2.show('net2.html')
```

```python
net3 = Network(notebook=True)
create_nodes_pyvis(net3, 100)
create_edges_pyvis(net3, 200)
net3.show('net3.html')
```

## Igraph

```python
from igraph import *
```

```python
def create_edges_igraph(graph, n=10):
    nodes = list(range(len(graph.vs())))
    nodes_lst = []
    if n > len(nodes)*(len(nodes)-1)/2:
        n = len(nodes)*(len(nodes)-1)/2
    while len(nodes_lst) != n:
        nodes1 = nodes.copy()
        i1 = random.choice(nodes1)
        nodes1.pop(i1)
        i2 = random.choice(nodes1)
        if (i1, i2) not in nodes_lst and (i2, i1) not in nodes_lst:
            nodes_lst.append((i1, i2))
    graph.add_edges(nodes_lst)
```

```python
graph1 = Graph()
graph1.add_vertices(10)
create_edges_igraph(graph1, 15)
plot(graph1, layout="kk", bbox = (500, 250))
```

```python
graph2 = Graph()
graph2.add_vertices(50)
create_edges_igraph(graph2, 60)
plot(graph2, layout="kk", bbox = (500, 500))
```

```python
graph3 = Graph()
graph3.add_vertices(100)
create_edges_igraph(graph3, 150)
plot(graph3, layout="kk", bbox = (700, 400))
```
